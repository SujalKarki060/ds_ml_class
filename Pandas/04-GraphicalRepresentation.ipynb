```python
import pandas as pd
```

```python
titanic_df = pd.read_csv("titanic_data.csv")
```

```python
titanic_df["Survived"].value_counts().plot(
    kind="bar",
    title = "Count of Survived Passengers",
    figsize = (4,4),
    legend = True,
    yticks = list(range(0,600,100)),
    xlabel = "Survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    colormap = "bwr"
)
```

```python
group_by_gender = titanic_df[["Sex","Survived"]].groupby(by = "Sex")
```

```python
pd.crosstab(titanic_df.Survived,titanic_df.Sex).plot(
    kind="bar",
    title = "Count of Survived Passengers",
    figsize = (4,4),
    legend = True,
    yticks = list(range(0,500,50)),
    xlabel = "Survived (0=No, 1=Yes)",
    ylabel = "No. of Passengers",
    grid = True,
    colormap = "Dark2"
)
```

```python
mean_age_by_embarked = titanic_df.groupby("Embarked")["Age"].mean()
mean_age_by_embarked
```

```python
mean_age_by_embarked.plot(
    kind="bar",
    title="Mean Age of Passengers by Embarked Port",
    figsize=(6, 4),
    legend=True,
    xlabel="Embarked (C = Cherbourg, Q = Queenstown, S = Southampton)",
    ylabel="Mean Age",
    colormap="bwr",
    yticks=list(range(0, 50, 5))
)
```

```python
median_age_by_embarked = titanic_df.groupby("Embarked")["Age"].median()
median_age_by_embarked.plot(
    kind="pie",
    subplots=True,
    legend=True
)
```

```python
pd.crosstab(titanic_df.Survived,titanic_df.Sex).plot(
    kind="pie",
    subplots= True,
    #title = "Count of Survived Passengers",
    fontsize=10,
    figsize = (8,8),
    legend = True,
    colormap = "Dark2"
)
```
